```python
axis_size = 15
val_size = 17
title_size = 20

list_of_files = [
                ("aiidadb_davide_dj_1m_p_md_query_raw.txt", "Max depth"),
                ("aiidadb_davide_dj_1m_p_mh_query_raw.txt", "Max height")
                ]
p_title = "Max height, Max depth popularity"
x_title = "Max depth/Max height"
y_title = "Occurences"

data_limit = 120

plt.rc('axes', titlesize=axis_size)     # fontsize of the axes title
plt.rc('axes', labelsize=axis_size)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=axis_size)    # fontsize of the tick labels
plt.rc('ytick', labelsize=axis_size)    # fontsize of the tick labels
plt.rc('legend', fontsize=axis_size)    # legend fontsize
```

```python
import pylab
import matplotlib.pyplot as plt
from scipy.interpolate import spline
import scipy.signal as signal
import pandas as pd


datalist = [ ( pylab.loadtxt(filename), label ) for filename, label in list_of_files ]

# plt.yscale('log')
for data, label in datalist:
    pylab.plot( data[:data_limit,0], data[:data_limit,1], label=label )

pylab.legend()
pylab.title(p_title, fontsize=title_size)
pylab.xlabel(x_title)
pylab.ylabel(y_title)
```

```python
import pylab
import matplotlib.pyplot as plt
from scipy.interpolate import spline
import scipy.signal as signal
import pandas as pd
from pandas import Series

datalist = [ ( pylab.loadtxt(filename), label ) for filename, label in list_of_files ]

plt.yscale('log')
for data, label in datalist:
    series = Series.from_array(data[:data_limit,1])
#     rolling = series.rolling(window=10)
    rolling = series.rolling(3, win_type='triang')
    rolling_mean = rolling.mean()
    pylab.plot( data[:data_limit,0], rolling_mean, label=label )

pylab.legend()
pylab.title(p_title, fontsize=title_size)
pylab.xlabel(x_title)
pylab.ylabel(y_title)
```
